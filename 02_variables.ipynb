```python
# sep="" 
# tells print — "put nothing between items instead of a space."

print("a", "b", "c")          # default → a b c   (space between)
print("a", "b", "c", sep="")  # no gap  → abc
print("a", "b", "c", sep="-") # custom  → a-b-c
```

```python
#Learning from Error

# Can Python do this? Try all of these and see what happens
print(10 + 5)       # int + int
print(10 + 5.0)     # int + float  
print("HDFC" + " Bank")   # string + string


#print("Price: " + 1520.50)  # string + float → what happens?
```

```python
#way1 - f-string
price = 1520.50
print(f"Price: {price}")          # Python handles the type for you
```

```python
#way2 - str() conversion
print("Price: " + str(1520.50))   # str() turns float into "1520.5"
```

```python
#way3 - comma in print
print("Price:", 1520.50)          # comma, not +
```

```python
# Type Conversion Summary:
# str(1520.50)  → "1520.5"   float to string
# str(100)      → "100"      int to string
# int("100")    → 100        string to int
# float("19.5") → 19.5       string to float

# TypeError happens when you mix types with +
# Fix: always use f-strings when combining text and numbers
```
