```python
# Comparison Operators
# compares two values and return True or False based on the result

print(10 == 10)
print(10 != 10)
print(7 > 3)
print(7 >= 3)
print(7 < 3)
print(7 <= 7)
```

```python
#strings can be compared too
#you can compare strings too alphabetically not just numbers

print('a' < 'b')
print('a' == 'b')
print('a' == 'A')

# = assigns 
# == compares
```

```python
# chained comparison - evaluates from left to right, checking each cond. one by one
print(5 < 4 < 6)

# Is age between 18 and 30
import random
age = random.randint(18,50)
print(age)
print(18 <= age <=30) #its like check a value is between two bounds
```

```python
# Logical Operators
# used to combine multiple boolean expressions
# and - both must be true
# or - atleast one is true

print(3 > 1 and 5 < 1)
print(3 > 1 and 5 > 1)

print(3 > 1 or 5 < 1)
print(3 < 1 or 5 < 1)
```

```python
#check if system is under pressure
cpu_uage = 70
memory_usuage = 65

print(cpu_uage > 90 or memory_usuage > 90)
```

```python
# checking user credentials before login

email = True
password = False

print(email and password)
```

```python
# Not Operator - reverse the truth (True to False, False to True)
print(not 3 > 2 )
print(not True )

name = ""
print(not name)
```

```python
# priority order - and, or, not

print(5 == 5 or 8 > 5 and 6 < 4)

# control the execution with () - simple parenthesis
print((5 == 5 or 8 > 5) and 6 < 4)
```

```python
# exercise
# allow access only if the user is logged in or guest but not banned

is_logged_in = True
is_guest = False
is_banned = True

#print(is_logged_in or is_guest and not is_banned) - not the order we want

print((is_logged_in or is_guest) and not is_banned) # we control the execution order
```
