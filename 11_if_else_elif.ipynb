```python
# Conditional Statements
# checkpoint that cheks a condition in code - like traffic light
# True? runs the code
# False? skip it

# standalone if  - check cond. if true, do something. if not, just skip it
import random
score = random.randint(50, 100)
print(score)
if score >= 90:
    print("A")  # identation - part of the syntax
else: #else does not need a condition
    print("F")
# identation - adding spaces at the beginning to show that line belongs to a code block
```

```python
#Multiple conditions
#elif - aska follow-up question only runs if previous cond. were false

import random
score = random.randint(0, 100)
submitted_project = True
print(score)
if score >= 90:
    print("A")  
elif score >= 80 <90:
    print("B")
elif score >= 70 <80:
    print("C")
elif score >= 60 <70:
    print("D")
else:
    print("F")
```

```python
# Nested if - if statement inside another if

score = 95
submitted_project = True
if score >= 90:
    if submitted_project:
        print("A+")
    else:
        print("A")  
```

```python
# connecting conditions 
score = 95
submitted_project = False
if score >= 90 and submitted_project:
    print("A+")
elif score >= 90:
    print("A")
```

```python
# independent if - each if is checked separatedly 
# all conditions are tested - even if one is already true

score = 95
submitted_project = False

if score >= 90:
    print("High Score")
else:
    print("Low Score")

if submitted_project:
    print("Project is submitted")
else:
    print("Project is not submitted")
```

```python
# inline if - Ternary - no elif

score = 95
Grade = "A" if score >= 90 else "F" # assign it to the variable 
print(Grade) # we can print the assigned variable
```

```python
# add more stuffs
score = 79
Grade = "A" if score >= 90 else "B" if score >= 80 else "F" #elif not allowed here
print(Grade)
```

```python
# case - match - evaluate a value against multiple values run the code of the first match

country = "USA"
if country == "United States":
    print("US")
elif country == "India":
    print("IN")
else:
    print("unknown country")

# Easy to read and write
# can be used only for matching values
match country:
    case "United States" | "USA":
        print("US")
    case "India":
        print("IN")
    case "Germany":
        print("DE")
    case _:
        print("unknown Country")
```

```python
# Challenge 1
# validate the quality and correctness of Email 
# must not be empty
# must contain "." and "@"
# must end with ".com", ".org", ".net"
# must not be longer than 254 characters
# must start and end with a letter or digit

email = "shiny@duck.com"  

#clean the data
emails = email.strip()

if emails == "":
    print("Email cannot be empty") #cond 1
elif ',' not in emails and '@' not in emails:
    print("Email must contains . and @") #cond 2
elif emails.count("@") != 1:
    print("Email must contain exactly one @")
elif not emails.endswith((".com",".org",".net")):
    print("Email must end with .com, .org, .net")
elif len(emails) > 254:
    print("Email must not be longer than 254 characters")
elif not(emails[0].isalnum() and emails[-1].isalnum()):
    print("Email must starts and ends with letter or digit")
else:
    print("Email is Valid")
```

```python
# challenge 2
# validate the quality and correctness of Password
# must not be empty
# atleast 8 character
# atleast 1 uppercase
# atleast 1 lowercase
# not same as email
# not contain any spaces
# starts and ends with letter or digit

email = "shiny@duck.com"
password = "Seesawd_"

# must not be empty
if password == "":
    print("Password must not be empty")

# atleast 8 characters
elif len(password) < 8:
    print("Password must be atleast 8 characters")

# atleast 1 uppercase
elif not any(char.isupper() for char in password):
    print("Password must include atleast 1 uppercase")

# atleast 1 lowercase
elif not any(char.islower() for char in password):
    print("Password must include atleast 1 lowercase")

# not same as email
elif password == email:
    print("Password must not be same as email")

# no spaces
elif " " in password:
    print("Password must not contain spaces")

# starts and ends with letter or digit
elif not (password[0].isalnum() and password[-1].isalnum()):
    print("Password must start and end with letter or digit")

else:
    print("Password is Valid ✅")
```
