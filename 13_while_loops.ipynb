```python
# while Loop
# rpeats a block of code - over and over as long as condition is True!

# for vs while:
# for - based on predefined sequence "known times", predefined conditions
# while - "unknown Times!" (Risk)
```

```python
# while condition
i = 1 # intialization 
while i < 6: # condition
    print(i)
    i += 1 #update the variable
```

```python
answer = ""
while answer != "yes":
    answer = input("Do you agree? (yes/no):")
print("Thank You")
```

```python
# while True - always true

while True:
    x = input("Type")
    if x == 'stop':
        break
```

```python
# For 
# loop over a fixed sequence
# predefined condition
# no of iteration is know
# processing Data
# simple, clean, safe, Limited

# while
# loop while a cond. is true
# your Condition
# no. of iteration is unknown
# "waiting for external event"
# advanced, flexible, complex and risk

```

```python
# unlimated challence to this code
# allow up to  3 attempts
# user types "yes", print "Glad we are on same page"
# other print "3 strikes, you are out!"

attempts = 0

while attempts < 3:
    answer = input("Do you agree? (yes/no): ")

    if answer.lower() == "yes":
        print("Glad we are on same page")
        break

    attempts += 1
else:
    print("3 strikes, you are out!")
```
