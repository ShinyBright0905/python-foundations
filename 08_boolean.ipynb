```python
# Logic & Operators
# Control the flow of the code through Conditional Statement or through Loops
# conditional Statement - if, else, elif
# Loops - for, while, break, pass, continue
```

```python
# statement alone not enough we need condition which are expressed with boolean expressions
# bool expressions - values, functions, comparison operators logtical, membership, identity

# values - True, False
# Functions - bool(), any(), all(), isinstance()
# comparison - ==, !=, <, >, <=, >=
# logical - and, or, not
# Membership - in, notin
# identity - is, is not
```

```python
#Boolean

print(True)
print(False)
print(type(True))

# True - if Value is non-empty or Non-zero
# False - if value is Empty or zero
print(bool(123))
print(bool("Hi"))
print(bool())
print(bool(0)) 
print(bool("")) #empty
print(bool(None)) #empty - nothing
```

```python
# any - returns True if atleast one value is True
# all - returns True if all values are True

email = "shiny@gmail.com"
phone = "0176-123456"
username = "shiny"

#allows registration - if any field is filled
print(any([email,phone,username]))

#allows registration - only if all field is filled
print(all([email,phone,username]))
```
