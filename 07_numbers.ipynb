```python
# 2_math operators
print(2 + 3)
print(5 - 3)
print(4 * 2)
print(7 / 2)    # 3.5  → float
print(6 / 2)    # 3.0  → still float, not 3
print(6 // 2)   # 3    → int, floor division
print(7 % 2) #remainder
print(7 ** 2) #exponentiation
```

```python
x = 2
#x = x + 3
x += 3 #shorthand assignment - apply operator and reassign the result in one step
print(x)

x -= 1
print(x)
```

```python
# 3_rounding
# abs() - returns the absolute non-negative value of a number - distance and size measuring
import math
print(abs(2 - 10))

# rounding numbers
# floor() - rounds a number down to the floor
# ceil() - rounds a number up to the ceiling - pages or batches 
# round() - rounds a number to the nearest whole number up or down depending on whats closer

price = 35.5432189
# rounds the numbers to the specified number of decimal places
print(round(price, 2))

# import math before using floor and ceil
print(math.floor(price))
print(math.ceil(price))

# trunc() - cuts off the decimal part and keep the whole number (no rounding)
print(math.trunc(price))
print(int(price))

# int() - not using math already, its simple
# trunc() - imported math
```

```python
# 4_advanced maths
import math

math.sqrt(144)      # √144 = 12.0  → std deviation calculations
math.log(1.08)      # natural log  → continuous compounding
math.log(100, 10)   # log base 10
math.pi             # 3.14159...
math.e              # 2.71828...   → Euler's number, used in finance
math.pow(2, 10)     # 2^10 = 1024  → same as ** but returns float
```

```python
# 6_validation
# is_integer() - check if numbers are truly whole (floats with .0)
x = 7.1
print(x.is_integer())

y = 7.0
print(y.is_integer())
```

```python
# is_instance - checks if a value belongs to a certain data type
x = 70.4
print(isinstance(x, int))
print(isinstance(x, float))
```

```python
# Challenge
# generate a random int between 1 and 100 and check the result is even

number = random.randint(1, 100)
remainder = number % 2
if remainder == 0:
    print(f"{number} is an even number")
else:
    print(f"{number} is not an even number")
```
