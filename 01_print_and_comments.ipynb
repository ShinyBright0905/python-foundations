```python
# Master Exercise
company = "HDFC Bank"   # string  → text, always in quotes
share_price = 1520.50   # float   → decimal number
shares_owned = 100      # integer → whole number
buy_price = 1200.00     # price i originally bought at

current_value = share_price * shares_owned
invested_value = buy_price * shares_owned
profit = current_value - invested_value


print(f"Company: {company}")
print(f"Invested: ₹ {invested_value:,.2f}")
print(f"Current Value: ₹ {current_value:,.2f}")
print(f"Profit: ₹ {profit:,.2f}")
```
