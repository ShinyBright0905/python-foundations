```python
# Copying Lists
# shallow copy - changes can be only on copy and not on original
letters = ['a','b','c']
copylst = letters.copy()
copylst.append('Z')
print(letters)
print(copylst)
```

```python
# Deep Copy - creates a true independent copy for all levels
import copy 
matrix = [['a', 'b'],  # Row 0
          ['c','d']]  # Row 1

matrix_copy = copy.deepcopy(matrix)
matrix.pop()
matrix_copy[0].append('z')
print(matrix)
print(matrix_copy)
```

```python
# IS Operator

import copy

original = [
    ['a','b'],
    ['c','d']
]

copy1 = original
print(original is copy1)

copy2 = original.copy()
print(original is copy2)
print(original[0] is copy2[0])

copy3 = copy.deepcopy(original)
print(original is copy3)
print(original[0] is copy3[0])
```
