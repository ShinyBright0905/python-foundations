```python
#Compound Interest calculation
#input Prinicipal, interest, period
principal_amount = float(input("Enter Principal Amount: "))
interest_rate = float(input("Enter Annual Interest Rate: "))
period = int(input("Enter Number of Years: "))

#simple interest calculation
simple_interest = (principal_amount * interest_rate * period)/100
total_amount = principal_amount + simple_interest

# compound interest calculation
compound_amount   = principal_amount * (1 + interest_rate/100) ** period
compound_interest = compound_amount - principal_amount

# comparison
print(f"\n--- Comparison ---")
print(f"Principal:         ₹ {principal_amount:,.2f}")
print(f"Simple Interest:   ₹ {simple_interest:,.2f}")
print(f"Compound Interest: ₹ {compound_interest:,.2f}")
print(f"CI earns extra:    ₹ {compound_interest - simple_interest:,.2f}")
```
