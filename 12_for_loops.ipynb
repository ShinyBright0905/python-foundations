```python
# for loop to control statement - break, continue and pass
# break statement - stops the loop immediately, ends the loop right away

# example - break statement
for i in (1,2,3):
    if i==3:
        break
    print(i)


names = ['john','maria','','kumar']
for name in names:
    if name == "":
        print("Empty value Detected!")
        break
    print(f'Name = {name}')
```

```python
# continue statement - skip one loop cycle without stopping the loop
# skip this one and go next, skip bad or empty data without stopping the loop 

# example - continue statement
for i in (1,2,3):
    if i==2:
        continue
    print(i)
```

```python
# with continue 
names = ['john','maria','','kumar']
for name in names:
    if name == "":
        print("Empty value Detected!")
        continue
    print(f'Name = {name}')
```

```python
# pass statement - nothing happens, just keep going do nothing
# its like a passholder
for i in (1,2,3):
    if i==2:
        pass
    print(i)
```

```python
# check whether any filename appears more than once
# print "Duplicate found" - if duplicate exists, otherwise print "All fiels are unique"

files_list = [
    'report.csv',
    'data.txt',
    'summary.docx',
    'report.pdf',
    'data.csv'
]

# extract just the names without extension
names_only = []
for file in files_list:
    name = file.split(".")[0]   # 'report.csv' → ['report','csv'] → 'report'
    names_only.append(name)

# now check for duplicates in names_only
found_duplicate = False
for name in names_only:
    if names_only.count(name) > 1:
        print(f"Duplicate name found: {name}")
        found_duplicate = True
        break

if not found_duplicate:
    print("All filenames are unique")
```
