```python
# membership (in) operator
# 'a' in "data"   - True
# 'a' not in "data"   - False

print("o" in "python")
print(3 not in [1,2,3])
```

```python
# security check: ensure the domain is not banned
domain = "gmail.com"
banned_domains = ["duck.com", "spam.com", "fake.com"]
print(domain not in banned_domains)
```

```python
# identity operators - checks if two variables refer to the same object in memory
a = [1,2,3,4]
b = [1,2,3,4]
print(a == b) #compare values - True
print(a is b) #compare ID - False

a = 5
b = 5
print(a == b) #compare values - True
print(a is b) #compare ID - True
```

```python
#between two variables - assign one variable to the same object that another variable referring to
a = [1,2,3,4]
b = a
print(a == b) #compare values - True
print(a is b) #compare ID - True
```
