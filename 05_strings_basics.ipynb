```python
#Categories
#1_Types
#Type() - returns the datatype of the variables
name = 'Shiny'
print(type(name))

age = 25
print(type(age))
print("Your age is:" + str(age)) #changed from int to str just for this print function not change his varaible to str
print(type(age))

#str()
age = str(age) #however through this new updated variable we changes its type from int to str
print(type(age))
```

```python
#2_Math 
#len() - returns number of items in a value 
password = ' 12365484a'  #len() counts everything even spaces
print(len(password))

if len(password) < 8:
    print("Your password is too Short!")


text = """
Python is easy to lear. 
Python is Powerful.
Many people love python"""

#count() - str method output: int (returns how often a word appears in the string)
print(text.count("Python")) #python is case senstitive
print(text.count("python")) #python is case senstitive
```

```python
#3_Transformations
#replace() - replace the old value with new value
price = "1234,56"
print(price.replace(",","."))

phone_number = "176-1234-56"
print(phone_number.replace("-", "/"))

print(phone_number.replace("-", "")) #remove the character

prices = "$1,299.99"
print(prices.replace("$", "").replace(",",""))
```

```python
#Exercise for clean number format
#input = +49 (176) 123-4567
#output = 00491761234567
number = "+49 (176) 123-4567"
print(number.replace("+","00").replace(" ","").replace("-","").replace("(","").replace(")",""))
```

```python
#Extraction
#Indexing 
#in string each character has a position number (index)
# +ve index - Left to right start with o
# -ve index - right to left start with -1

#slicing
#extract more character with [start:end]
#start - included 
#end - not included 

#open-ended string - no specific end [1: ] 

#[Start:end:Step]
#[0:4:2] - skiping every 2nd character 
```

```python
# exercise of Indexes and Slicing
Python = "Python"

# extract first character
print(Python[0])
print(Python[-6])

# Extract last character
print(Python[5])
print(Python[-1])

# extract h
print(Python[3])
print(Python[-3])

date = "2026-09-20"

# extract year
print(date[0:4])
print(date[:4])

# extract the month
print(date[5:7])

# extract the date
print(date[8:])
print(date[-2:])
```
