```python
# Iterables & Filtering
# Iterable - anything you can loop over (list, str, tuple, dict, range...)
# Python has built-in tools to loop, transform and filter iterables
# without writing a manual for-loop every time: enumerate(), reversed(),
# zip(), map(), filter()

# Manual way - looping and building a new list ourselves
letters = ['a', 'b', 'c', 'd', 'e']
new_list = []
for l in letters: 
    new_list.append(l.upper())
print(new_list)
```

```python
# enumerate() function adds a counter to an iterable and returns it in a form of enumerate object. This enumerate object can then be used directly in for loops or be converted into a list of tuples using list() method.
letters = ['a', 'b', 'c', 'd', 'e']

for i, l in enumerate(letters, start=1):
    print(i, l)
```

```python
# reversed() function returns a reversed iterator. It can be used to iterate over a sequence in reverse order.

print(list(reversed(letters)))
```

```python
# zip() function takes iterables (can be zero or more), aggregates them in a tuple, and returns it. This function is used to combine two or more iterables (e.g. lists, tuples) element-wise.
numbers = [1, 2, 3, 4, 5]
# print(list(zip(letters, numbers)))


for l, n in zip(letters, numbers):
    print(l, n)
```

```python
# Transforming iterables using map() function
letters = ['a', 'b', 'c', 'd', 'e']

map(str.upper, letters)  # returns a map object
print(list(map(str.upper, letters)))  # returns a list of uppercase letters
```

```python
# map() is lazy - it doesn't run until you consume it (list(), for-loop, etc.)
numbers = ['1', '2', '3', '4', '5']
map(int, numbers)  # returns a map object
print(list(map(int, numbers)))  # returns a list of integers
```

```python
name = ['  maria', '  john', ' alice']
map(str.strip, name)  # returns a map object
# print(list(map(str.strip, name)))  # returns a list of stripped names

for n in map(str.strip, name):
    print(n)
```

```python
letters = ['a', '10', '', None, False]

print(list(filter(None, letters)))  # returns a list of non-empty strings
#print(list(filter(bool, letters)))  # returns a list of truthy values
```

```python
items = ['sql', '123', 'python', '456', 'java']
filter(str.isalpha, items)  # returns a filter object
# print(list(filter(str.isalpha, items)))  # returns a list of strings that contain only
for i in filter(str.isalpha, items):
    print(i)
```

```python
# Iterables & Filtering Summary:
# enumerate()  -> adds a counter to an iterable
# reversed()   -> reverses the order (wrap in list() to see it)
# zip()        -> combines multiple iterables element-wise
# map()        -> transforms every item using a function
# filter()     -> keeps only items where the function/condition is True
# All of these return lazy objects - wrap in list() or loop over them to see results
```
