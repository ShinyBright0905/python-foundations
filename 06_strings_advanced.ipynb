```python
#New Challenge
#data - 968-Maria, ( D@t@ Engineer  );; 27y  
#output - Name: maria role: data engineer age: 27

data = "968-Maria, ( D@t@ Engineer  );; 27y  "

# Step 1 - first split by comma
parts = data.split(",")
# parts[0] = '968-Maria'
# parts[1] = ' ( D@t@ Engineer  );; 27y  '

# Step 2 - get name from parts[0]
name = parts[0].split("-")[1].lower()

# Step 3 - split parts[1] by ;; to separate role and age
parts2 = parts[1].split(";;")
# parts2[0] = ' ( D@t@ Engineer  )'
# parts2[1] = ' 27y  '

# Step 4 - clean role
role = parts2[0].strip().replace("(","").replace(")","").replace("@","a").strip().lower()

# Step 5 - clean age
age = parts2[1].strip().replace("y","")

print(f"Name: {name} Role: {role} Age: {age}")
```

```python
# 5_Search
# searching
# startswith() - methods - check if string starts with specific word #bool results
# endswith() - methods - checks if string ends with a specific word #bool results
# in - checks if a word exists in the string - output - bool

phone = "+49-176-12345"
phone1 = "+48-657-98745"
phone2 = "+0048-784-46518"
print(phone.startswith("+49"))

email = "shiny@outlook.com"
print("@" in email)  # to valid the email
print(email.endswith("@gmail.com"))

file = "data.csv"
print(file.endswith(".csv"))

# find function - returns the starting position of a word in the string - output - number
print(phone[phone.find("-") + 1:])
print(phone1[phone1.find("-") + 1:])
print(phone2[phone2.find("-") + 1:])
```
